```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.cluster import KMeans
```

```python
#Übung 1
data = np.loadtxt('data/clicks.dat', delimiter=',')
model = KMeans(n_clusters=5, max_iter=200) #Anzahl Cluster und Iteration (Anzahl bis Abbruch) anpassen
y_pred = model.fit_predict(data) #Vorhersage bestimmen
```

```python
print(model.cluster_centers_) #Anzeige von Koordinaten-Mitte (Cluster-Center)
```

```python
print(model.inertia_)  #Trägheit des Clusters
```

```python
#Visualisierung
col = ListedColormap(['red', 'blue', 'green', 'cyan', 'yellow'])
sp = plt.scatter(data[:,0], data[:,1], c=y_pred, cmap=col)
#Legende
lab = (['C1', 'C2', 'C3', 'C4', 'C5'])  #bei mehr clustern legendennamen & farben ergänzen
plt.legend(handles=sp.legend_elements()[0], labels=lab)
plt.grid(True)
plt.show()
```

```python
#Ellbogen-Methode
data = np.loadtxt('data/clicks.dat', delimiter=',')
inert = []
for k in range(1,11): #Anzahl k anpassen, wird für jedes k in Range durchgeführt
  model = KMeans(n_clusters=k)
  model.fit(data)
  inert.append(model.inertia_)
```

```python
#Visualisierung
x = np.linspace(1, 10, 10)
plt.xlabel(f'$k$')
plt.ylabel(f'Trägheit')
plt.plot(x, inert, 'b-')
plt.plot(x, inert, 'bo')
plt.show()
```
