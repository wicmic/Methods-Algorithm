```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split as tts
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
varianz = [0., 80., 18., 1., 0.25, 0.25, 0.25, 0.25, 0.]
x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
plt.xlabel("Hauptkomponenten")
plt.ylabel("Varianz [%]")
plt.plot(x, np.cumsum(varianz), "r-")
plt.show()
```

```python
#PCA
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data
y_true = iris.target
iris
```

```python
from sklearn.decomposition import PCA
model_pca = PCA(n_components=2)
data_proj = model_pca.fit_transform(x)
print(model_pca.explained_variance_ratio_) #hier: 92.4% entlang der ersten Hauptkomponente, 5.3% entlang der zweiten Hauptkomponente --> 97.8%
plt.scatter(x[:,0], x[:,1], c=y_true)
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data
y_true = iris.target
from sklearn.decomposition import PCA
model_pca2 = PCA(2)
data_proj = model_pca2.fit_transform(x)
plt.scatter(x[:,0], x[:,1], c=y_true)
plt.show()
```

```python
model_pca2.explained_variance_ratio_
```

```python
from sklearn.cluster import KMeans
model_km = KMeans(n_clusters=3) #aus Abbildung oben
y_pred = model_km.fit_predict(data_proj) #Vorhersage bestimmen

plt.scatter(x[:,0], x[:,1], c=y_pred)
plt.show()
```
