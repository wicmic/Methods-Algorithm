```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons
```

```python
#Daten darstellen
data2, y_true = make_moons(n_samples=300, noise=0.05)  #data2 ist Datensatz, y_true echte Klassifizierung
plt.scatter(data2[:,0], data2[:,1], c=y_true)
```

```python
inert = []
for k in range(1,11):
  model = KMeans(n_clusters=k)
  model.fit(data2)                  #Datensatz korrekt benennen
  inert.append(model.inertia_)
#Visualisierung
x = np.linspace(1, 10, 10)
plt.xlabel(f'$k$')
plt.ylabel(f'Trägheit')
plt.plot(x, inert, 'b-')
plt.plot(x, inert, 'bo')
plt.show()
```

```python
#k-means
model = KMeans(n_clusters=4, max_iter=200)  #Anzahl Cluster von Knick aus Ellbogen (hier 2, könnte aber auch 4 sein als Alternative)
y_pred = model.fit_predict(data2)
```

```python
#Visualisierung
col = ListedColormap(['red', 'blue', 'green', 'cyan', 'yellow'])
sp = plt.scatter(data2[:,0], data2[:,1], c=y_pred, cmap=col)
#Legende
lab = (['C1', 'C2', 'C3', 'C4', 'C5'])  #bei mehr clustern legendennamen & farben ergänzen
plt.legend(handles=sp.legend_elements()[0], labels=lab)
plt.grid(True)
plt.show()
```
