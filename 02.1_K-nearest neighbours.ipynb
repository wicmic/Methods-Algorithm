```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score as score
```

```python
data = np.loadtxt('data/credit.dat', delimiter=',')
x_train, x_test, y_train, y_test = tts(data[:,:2], data[:,2], test_size=0.3) #test_size Anteil Testf√§lle % anpassen
model = knn(n_neighbors=7) #Anzahl knn anpassen
model.fit(x_train, y_train)
y_pred = model.predict(x_test) #Vorhersage bestimmen
print(score(y_test, y_pred))
```

```python
#Visualisierung
col_train = ListedColormap(['red', 'blue'])
col_test = ListedColormap(['orange', 'cyan'])
sp = plt.scatter(x_train[:,0], x_train[:,1], c=y_train, cmap=col_train)
plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap=col_test)
#Legende
lab = ['niedrig', 'hoch']
plt.legend(handles=sp.legend_elements()[0], labels=lab)
plt.grid(True)
plt.show()
```

```python
#√úbung 2
data = np.loadtxt('data/credit.dat', delimiter=',')
x_train, x_test, y_train, y_test = tts(data[:,:2], data[:,2], test_size=0.3) #test_size Anteil Testf√§lle % anpassen
for k in range(1,16): #Anzahl k anpassen, wird f√ºr jedes k in Range durchgef√ºhrt
  model = knn(n_neighbors=k)
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test) #Vorhersage bestimmen
  print(k, ':', score(y_test, y_pred))
```

```python
#Visualisierung
col_train = ListedColormap(['red', 'blue'])
col_test = ListedColormap(['orange', 'cyan'])
sp = plt.scatter(x_train[:,0], x_train[:,1], c=y_train, cmap=col_train)
plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap=col_test)
#Legende
lab = ['niedrig', 'hoch']
plt.legend(handles=sp.legend_elements()[0], labels=lab)
plt.grid(True)
plt.show()
```
